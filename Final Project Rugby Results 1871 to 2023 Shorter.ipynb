### History

Rugby as a sport began in early 19th century. Legend has it that in 1823, a young man named William Webb Ellis picked up the ball during a football game and ran with it, creating a new game in the process, although historians question the authenticity of this story. Nevertheless the popularity of the new sport increased, with the first international match taking place between England and Scotland in 1871. The sport continued it's spread and in 1883, the Home Nations (England, Scotland, Wales, and Ireland) began an annual competition known as the Home Nations Championship, which later evolved into the Five Nations Championship after the addition of France in 1910. It became the Six Nations Championship in 2000 when Italy joined.

Rugby Union tends to be divided geographically by northern and southern hemispheres, outside of the northern hemisphere the sport took hold in New Zealand, Australia and South Africa. These countries would themselves create a yearly contested competition now known as The Rugby Championship. More recently Argentina have joined their ranks, with Japan improving significantly after hosting the 2019 World Cup. 

The Rugby World Cup, is one of the sport's most significant events and was first held in 1987. The tournemant features teams from around the world competing for the Webb Ellis Cup. So far only New Zealand, Australia, South Africa and England have tasted success by winning the tournement. 


Interest to note is that Rugby Union has been predominantly amateur throughout much of its history, with players pursuing professional careers outside of the sport. This amateur status eventually led to a dispute and the establishment of rugby league in 1895, a separate code of the sport with distinct rules. Due to the lack of available data, rugby league will not be analysed in this report.

The amateur status of rugby union changed in August 1995 when the International Rugby Board (IRB) allowed for the full-time payment of rugby union players for the first time. England international Mike Catt became the first fully professional rugby union player.

### Scoring

There are various methods of scoring in Rugby Union, both during gameplay when the ball is contested by the two teams, and during moments when the ball is not in play, such as penalty kicks.

Here is an overview of the terminology and the current ways in which scoring can occur. Understanding these concepts will enhance your ability to follow the game, as it provides insights into the objectives of both the attacking and defending teams and the sometimes unexpected allocation of points.


__Try:__

- When the ball is touched down by a player of a team in their opposition's in-goal area (on or behind the opposition's goal posts as in the diagram above).
- Currently worth 5 points.
- The highest but usually the most difficult method of scoring points.

__Conversion:__

- After a try is scored, the scoring team has an opportunity to kick the ball from the floor at their opponent's set of goal posts (situated on the goal line of their in-goal area).
- In order to score points, the ball must pass between the two vertical posts and above the horizontal post.
- The kick must be taken in front of where the try was scored, in the direction of the halfway line.
- If the team scores in the middle of the in-goal area, the conversion will be easier.
- Worth 2 points.
- After scoring a try, the maximum points available is 7.

__Penalty:__

- Given for certain types of infringements.
- Examples of infringements include contact with the head of an opponent and tripping with the feet.
- The team benefiting from the penalty has the option to kick for the touchline for a lineout (a form of contested throw-in) where the ball crossed the touchline, or take a penalty kick at goal from where the infringement took place.

__Penalty kick at goal:__

- Similar to a conversion.
- An attempt to kick the ball through the opponent's goal posts from the place of the infringement.
- Worth 3 points.
- This is a common way of scoring if a team has a skilled kicker (which is the case for most top international sides).

__Drop goal:__
- At any time during play, the team in possession of the ball can attempt to kick the ball from their hands through the opponents goal posts.
- The ball must bounce off the floor before being kicked by the player for the drop goal to be valid.
- To score points, the ball must pass between the two vertical posts and above the horizontal post.
- Drop goals are typically attempted for last-minute points when scores are level or within 2 points.

### Data Wrangling and Additional Columns

Firstly I will amend the data to give extra columns for the winning margin (the amount by which the winning team won), the winning team and the losing team. These will be useful later when analysing scores over time and when looking at the most successful teams in the dataset. A binary variable for a home win is added as well as formatting of the date column as datetime rather than a string.


```python
results_df['Winning Margin'] = (results_df['home_score'] -
                                results_df['away_score']).abs()
results_df['Winning Team'] = (np.where(
    results_df['home_score'] > results_df['away_score'],
    results_df['home_team'],
    np.where(results_df['home_score'] == results_df['away_score'], 'Draw',
             results_df['away_team'])))

results_df['Losing Team'] = np.where(
    results_df['Winning Team'] == results_df['home_team'],
    results_df['away_team'], results_df['home_team'])

results_df['Losing Team'] = np.where(results_df['Winning Team'] == 'Draw',
                                     'Draw', results_df['Losing Team'])

results_df['home_win'] = np.where(
    results_df['Winning Team'] == results_df['home_team'], True, False)
results_df[
    'points_per_game'] = results_df['home_score'] + results_df['away_score']

results_df['date'] = pd.to_datetime(results_df['date'], format='%Y/%m/%d')
```

## Rugby Union Scores over time

```python
import matplotlib.dates as mdates


avg_per_year = results_df.groupby(pd.PeriodIndex(results_df['date'], freq="Y"))[['Winning Margin','home_score', 'away_score']].mean()
avg_per_year = pd.DataFrame(avg_per_year).reset_index()
avg_per_year['date']= avg_per_year['date'].dt.to_timestamp()
avg_per_year['mdate'] = avg_per_year['date'].apply(mdates.date2num)
```

In this section, I analyse the average winning margins, as well as home and away scores per year. The resulting averages are plotted on the line graphs below to address the following questions:
- How have scoring patterns changed over time?
- Have scoring rule changes impacted winning margins?

In general, an upward trend in scoring over time, both in terms of team scores and winning margins is observed. This aligns with the aforementioned dates regarding scoring changes. Therefore, scoring patterns and winning margins have generally increased over time, and it seems that scoring patterns are the main driver behind this trend. However, based on this data, I cannot determine whether the increased emphasis on tries in recent scoring rule changes has resulted in more tries being scored. It should be noted that the increase in the value of a try has typically been followed by an increase in the value of kicks at goal. Since a breakdown of the points scored in each match isn't available, making a conclusive judgement is not possible. Nonetheless, an increase in points per game generally implies more action per game, suggesting that matches may have become more captivating over time.

Additionally, it is challenging to differentiate whether the change in scoring system or rugby's shift to professionalism led to the substantial rise in scores in the late 1990s. ## The best international teams over the history of the sport

In this section, the analysis focuses on the success of teams in the professional era in order to identify the top-performing teams and potential exciting matchups for the upcoming World Cup. To begin, a new dataframe is created to calculate the wins, losses, and draws of each team, along with their win percentage. Although the code used for this calculation may be somewhat inefficient, it yields valuable insights for comparing teams. It is also used in the later comparisons between teams of interest.


```python
filtered_results = results_df[results_df['Winning Team'] != 'Draw'].copy()
filtered_results['date'] = pd.to_datetime(filtered_results['date'])

wins_year = filtered_results.groupby(
    [filtered_results['date'].dt.year, 'Winning Team']).size().reset_index(
        name='Wins per Year').rename(columns={'Winning Team': 'Team'})
losses_year = filtered_results.groupby(
    [filtered_results['date'].dt.year, 'Losing Team']).size().reset_index(
        name='Losses per Year').rename(columns={'Losing Team': 'Team'})

draws_only = results_df[results_df['Winning Team'] == 'Draw'].copy()
draws_year = pd.concat([
    draws_only.groupby([draws_only['date'].dt.year, 'home_team'
                        ]).size().reset_index(name='Draws per Year').rename(
                            columns={'home_team': 'Team'}),
    draws_only.groupby([draws_only['date'].dt.year, 'away_team'
                        ]).size().reset_index(name='Draws per Year').rename(
                            columns={'away_team': 'Team'})
]).groupby(['date', 'Team'])['Draws per Year'].sum().reset_index()

wins_losses = pd.merge(wins_year,
                       losses_year,
                       how='outer',
                       on=['date', 'Team'])
wins_losses = pd.merge(wins_losses,
                       draws_year,
                       how='outer',
                       on=['date', 'Team'])
wins_losses.fillna(0, inplace=True)

wins_losses['Games per Year'] = wins_losses['Wins per Year'] + wins_losses[
    'Losses per Year'] + wins_losses['Draws per Year']
wins_losses['Win Percentage'] = wins_losses['Wins per Year'] / wins_losses[
    'Games per Year']

wins_losses['date'] = pd.to_datetime(wins_losses['date'], format='%Y')
win_loss_pro = wins_losses[wins_losses['date'] > "1995-08-01"]

group_pro = win_loss_pro.groupby('Team').sum()
group_pro['Win Percentage'] = (group_pro['Wins per Year'] /
                               group_pro['Games per Year'] * 100).round(2)

group_pro = group_pro.rename(
    columns={
        'Wins per Year': 'Total Wins',
        'Losses per Year': 'Total Losses',
        'Draws per Year': 'Total Draws',
        'Games per Year': 'Total Games'
    })
group_pro = group_pro.sort_values(by='Win Percentage',
                                  ascending=False).reset_index()

wc_df = results_df[results_df['Competition Series'] == 'World Cup']
filtered_results = wc_df[wc_df['Winning Team'] != 'Draw'].copy()
filtered_results['date'] = pd.to_datetime(filtered_results['date'])

wins_losses_year = pd.concat([
    filtered_results.groupby(
        [filtered_results['date'].dt.year, 'Winning Team']).size().reset_index(
            name='Wins per Year').rename(columns={'Winning Team': 'Team'}),
    filtered_results.groupby(
        [filtered_results['date'].dt.year, 'Losing Team']).size().reset_index(
            name='Losses per Year').rename(columns={'Losing Team': 'Team'})
])

draws_only = wc_df[wc_df['Winning Team'] == 'Draw'].copy()
draws_year = pd.concat([
    draws_only.groupby([draws_only['date'].dt.year, 'home_team'
                        ]).size().reset_index(name='Draws per Year').rename(
                            columns={'home_team': 'Team'}),
    draws_only.groupby([draws_only['date'].dt.year, 'away_team'
                        ]).size().reset_index(name='Draws per Year').rename(
                            columns={'away_team': 'Team'})
]).groupby(['date', 'Team'])['Draws per Year'].sum().reset_index()

wins_losses = pd.merge(wins_losses_year,
                       draws_year,
                       how='outer',
                       on=['date', 'Team'])
wins_losses.fillna(0, inplace=True)

wins_losses['Games per Year'] = wins_losses['Wins per Year'] + wins_losses[
    'Losses per Year'] + wins_losses['Draws per Year']

group_wc = wins_losses.groupby('Team').sum()
group_wc['Win Percentage'] = (group_wc['Wins per Year'] /
                              group_wc['Games per Year'] * 100).round(2)

group_wc = group_wc.rename(
    columns={
        'Wins per Year': 'Total Wins',
        'Losses per Year': 'Total Losses',
        'Draws per Year': 'Total Draws',
        'Games per Year': 'Total Games'
    })
group_wc = group_wc.sort_values(by='Win Percentage',
                                ascending=False).reset_index()

group_pro['df'] = 'Professional Era'
group_wc['df'] = 'World Cups Only'
group = pd.concat([group_pro, group_wc])
```

### Poisson Distribution and the Prediction Model

In this section, I'll be using the available dataset to make predictions for matches in the 2023 World Cup. It's time to get a bit technical as I explain the statistical method used. However, if you're not interested in the technical details, feel free to skip ahead to the predictions in the next section.

The Poisson distribution is a way to model the probability of a certain number of events occurring within a fixed interval of time or space that was briefly touched upon in the Codecademy Data Analytics. Using the Poisson distribution, the likelihood of different outcomes can be estimated, and predictions can be made based on the average rate of events. It is a handy tool in various fields, including statistics, mathematics, and even sports analytics. 

It relies on four key assumptions that must be met ([source](https://towardsdatascience.com/predicting-the-fifa-world-cup-2022-with-a-simple-model-using-python-6b34bdd4f2a5)):
1. The number of events can be counted.
2. The occurrence of events is independent.
3. The rate at which events occur is constant.
4. Two events cannot occur at exactly the same instant in time.

In the case of rugby, assumptions 1 and 4 are quite clearly met. The number of points scored by each team is a countable event, and since there's only one ball, teams can only score once at a time. While some arguments can be made about teams scoring leading to more scoring opportunities or tiredness affecting scoring probabilities towards the end of a game, for now, as the source suggests, the report will assume that assumptions 2 and 3 are met, .



```python
prof = results_df[results_df['date'] >="1995-01-01"]

def poisson(k,exp_events):
    minutes=80
    lam =(exp_events)
    p_k= np.exp(-lam)*np.power(lam,k)/factorial(k)
    return p_k


k_away=[]
p_k_away=[]
for i in range((prof['away_score'].max())):
    p_k_away.append(poisson(i,prof['away_score'].mean())*100)
    k_away.append(i)
    
dist_away = []

for prob in p_k_away:
    dist_away.append((prob/100)*len(prof))
    

k_home=[]
p_k_home=[]
for i in range(prof['home_score'].max()):
    p_k_home.append(poisson(i,prof['home_score'].mean())*100)
    k_home.append(i)
    
dist_home = []

for prob in p_k_home:
    dist_home.append((prob/100)*len(prof))
```

The function "poisson" (taken from [source](https://www.analyticsvidhya.com/blog/2022/06/predicting-the-scoreline-of-a-football-match-using-poisson-distribution/)) calculates the probability of observing a specific number of events (denoted by "k") within a given interval, assuming a Poisson distribution. Using this function, I calculate the probabilities based on the maximums and medians of the home and away scores in the data. Then, I plot these probabilities as a curve, calculate the expected distribution based on that curve, and compare it with the actual distribution from the dataset using a histogram.

As you can see, the expected distribution roughly matches the actual distribution, although not perfectly. This suggests that using the Poisson distribution to predict results may provide some level of accuracy, but there are likely better models that can fit the data.
```python
def PredictScore(home_team, away_team, data, neutral):

    matches = data[((data['home_team'] == home_team) &
                    (data.away_team == away_team)) |
                   ((data['home_team'] == away_team) &
                    (data.away_team == home_team))]
    recent_data = matches[(matches.date >= "2019-12-31")]
    matches = matches[(matches.date < "2019-12-31")]


    if neutral == False:
        avg_home_score = matches[(
            matches['home_team'] == home_team)].home_score.mean()
        avg_away_score = matches[(
            matches.away_team == away_team)].away_score.mean()

        recent_home_score = recent_data[(
            recent_data['home_team'] == home_team)].home_score.mean()
        recent_home_score = recent_home_score if not np.isnan(
            recent_home_score) else 0

        recent_away_score = recent_data[(
            recent_data.away_team == away_team)].away_score.mean()
        recent_away_score = recent_away_score if not np.isnan(
            recent_away_score) else 0

    else:
        avg_home_score = (
            matches[(matches['home_team'] == home_team)].home_score.sum() +
            matches[(matches.away_team
                     == home_team)].away_score.sum()) / len(matches)

        avg_away_score = (
            matches[(matches.away_team == away_team)].away_score.sum() +
            matches[(matches['home_team']
                     == away_team)].home_score.sum()) / len(matches)

        recent_home_score = (
            recent_data[(recent_data['home_team'] == home_team)].home_score.sum()
            + recent_data[(recent_data.away_team
                           == home_team)].away_score.sum()) / len(recent_data)

        recent_away_score = (
            recent_data[(recent_data.away_team == away_team)].away_score.sum()
            + recent_data[(recent_data['home_team']
                           == away_team)].home_score.sum()) / len(recent_data)

    weight_factor = 0.8

    if recent_home_score != 0 or recent_away_score != 0:
        avg_home_score = (weight_factor * recent_home_score) + (
            (1 - weight_factor) * avg_home_score)
        avg_away_score = (weight_factor * recent_away_score) + (
            (1 - weight_factor) * avg_away_score)

    home_points = int(stats.mode(np.random.poisson(avg_home_score, 10000))[0])
    away_points = int(stats.mode(np.random.poisson(avg_away_score, 10000))[0])

    if home_points > away_points:
        print(f'The winner will be {home_team}')
    else:
        print(f'The winner will be {away_team}')

    print(
        f'The scoreline will be {home_team} {home_points}:{away_points} {away_team}'
    )
```

This function takes inspiration from the approach outlined in the analytics vidhya blog ([here](https://www.analyticsvidhya.com/blog/2022/06/predicting-the-scoreline-of-a-football-match-using-poisson-distribution/)) mentioned earlier. It requires the following inputs: home team, away team, dataset, and a dummy variable (True or False) indicating whether the match is taking place at a neutral venue.

Here is a breakdown of the function's steps:

1. __Data splitting:__ The function splits the dataset into two parts: one containing results from the past 4 years (the most recent World Cup cycle) and the other containing results from before that period. This allows for a separate consideration of recent performance and historical data.


2. __Team average scores:__ If the dummy variable is set to False (indicating a non-neutral venue match), the function calculates average scores specifically between the home and away teams. However, if the match is at a neutral venue, all matches between the two teams are considered.


3. __Weighting:__ The function calculates two sets of average scores per team: one based on the most recent matches and another based on matches occurring more than 4 years ago. The function assigns higher weight to the recent scores to capture the teams' recent form. If no matches are found in the recent data analysis, the weighting step is skipped.


4. __Poisson distribution sampling:__ The function generates 10,000 random samples from a Poisson distribution using the average home and away scores obtained earlier. These samples represent the predicted scores for both teams.


5. __Determining the winner:__ The function identifies the most common value (mode) among the generated samples for both teams. The team with the higher predicted score is considered the winner.

The function considers recent form by giving it more weight in the predictions. This is important because it helps capture teams' recent performances. For example, as shown in Section 5 New Zealand have recently experienced a decline in form, while Ireland have been performing strongly leading up to the World Cup, the function aims to reflect these trends in the predictions.

I must acknowledge that this is a fairly simple way to make predictions, and in some cases too simple. In future I would like to learn more about predictions using machine learning, but for the time being I am applying what I have learnt in the Data Analytics course.

Nonetheless, the function strives to provide predictions that consider recent form and historical data to the best extent possible.
